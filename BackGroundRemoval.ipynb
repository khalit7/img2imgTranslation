```
import numpy as np
from PIL import Image
import glob
from os import listdir,remove
```

**Remove Gray scale images**

```
path_to_img="/content/drive/My Drive/grad_project/approach2/NoBackground_Dataset/NoBackground_5/"
```

```
import os
files = os.listdir(path_to_img)
print(len(files))
```

```
lst_img = glob.glob(path_to_img+"*.jpg")
counter=0
for i in lst_img:
  img = Image.open(i)
  img = np.array(img)
  if(len(img.shape) < 3):
    counter+=1
    print("done ",counter," times" ,flush=True)
    remove(i)
    print(i)
```
